```python
#First let us make a node class with a constructor just to reduce the reusing of code.

class Node(object):
    #let us define a constructor
    def __init__(self,data):
        self.data=data
        self.next=None
        
        
        
#let us define a Linkedlist class for root Node(head)        
class LinkedList(object):
    def __init__(self):
        self.head=None
        self.size=0
#Let us insert a element at begining of LinkedList

    def insertStart(self,data):
        self.size+=1
                                                # Increment the size of Linked List by 1
        newNode = Node(data)                    # Create a new Node with input "data"
        if(self.head!=None):                       # If head = NULL i.e it is first node in Linked List
            self.head = newNode                 # new value "data" assigned to new Node
        else:                                   # If head != Null i.e this is not the first node
            newNode.next = self.head        # Point the Pointer of "New Node" to current "Head Node"
            self.head = newNode                 # and make the "New Node" the "Head Node"
        return 'The Node {} added to Head of Linked List'.format(data)
#TO get the size of linked list let us define a constructor

    def linkedlistsize(self):
        return self.size
#let us insert a Node at the end of the Linked list    
    def insertEnd(self,data):
        self.size+=1                           #increment the size of Linked list by 1
        newNode=Node(data)                     #adding New Node with data 
        present_node=self.head                 #changing currentnode to head node
        while present_node.next is not None:   #checking if next node exists
            present_node=present_node.next     #traversing till next element of current node is NULL
        present_node.next=newNode              #adding the requried node at the end 
        return 'The Node {} is added to the End of the Linked List'.format(data)
            
#Let the Remove Node from Linked List        
    def RemoveNode(self,data):
        self.size-=1                           #Decerementing the size of Linked list by 1
        if (self.head==None):  
            return None
        currentNode=self.head                  #starting at the head and traversing for the required data to be removed
        prevNode=None
        while(currentNode.data!=data):
            prevNode=currentNode
            currentNode=currentNode.next
        
        if prevNode is None:                   # If the  Node to be deleted is the starting Node 
            self.head = currentNode.next
        else:
            prevNode.next = currentNode.next   #updating previous node to directly point to the next of current node(ignoring current node)
        return 'The Node {} is remove from the Linked List'.format(data)
    
    
#let us Traverse through the linked list 

    def TraverseList(self):
        currentNode=self.head
        while(currentNode!=None):              
            print('traversed node %d' % currentNode.data)
            currentNode=currentNode.next
            if currentNode.next==None:          #end of linkedlist so nothing to traverse
                return 'traversing done.No more elements to traverse'
            
```
