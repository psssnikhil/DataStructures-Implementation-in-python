```python
#initialize array with name arr

arr=[10,20,30,40,50]

#In python indexing is a O(1) operation.
print(arr[0])   #------> prints element at index '0'
#output is 100

#INSERT ITEM

arr[2]=100
print(arr[2])
print(arr)  #------> here the index 2 is replaced with 100 and is printed
#output [10, 20, 100, 40, 50]



#Now lets print all items in an array

for items in arr:
    print(items)
    
#now let us print with index

for i in range(len(arr)):
    print(i)
    print(arr[i])                #----->prints indexes as well as values 
    
    
#print items from one to another(start to end index)    
print(arr[1:4])     #-----------> 1 represents starting index and 4 reperesents end index'

#now lets index from the ending
print(arr[:-2])                 #------->  print all elements ignoring the last two indexes
                                #--------> can be used where in an array all except last one or two are needed.


#Now let us searh for an element in an array , it has time complexity of O(n)

a=arr[2]
print(a)

for i in arr:
    print(i)
    if i==a:
        print('found',i)
    else:
        print('not found')
    
```

###### Arrays implementation using Class and instances.#####

```python
class Array(object):
    
    ''' sizeOfArray: denotes the total size of the array to be initialized
       arrayType: denotes the data type of the array(as all the elements of the array have same data type)
       arrayItems: values at each position of array
    '''
    
    def __init__(self, sizeOfArray, arrayType):
        self.sizeOfArray = len(list(map(arrayType, range(sizeOfArray))))
        self.arrayItems =[arrayType(0)] * sizeOfArray    # initialize array with zeroes
        self.arrayType = arrayType

    def __str__(self):
        return ' '.join([str(i) for i in self.arrayItems])

    def __len__(self):
        return len(self.arrayItems)

    # magic methods to enable indexing
    def __setitem__(self, index, data):
        self.arrayItems[index] = data

    def __getitem__(self, index):
        return self.arrayItems[index]

    # function for search
    def search(self, keyToSearch):
        for i in range(self.sizeOfArray):
            if (self.arrayItems[i] == keyToSearch):      
                return i                                 

        return -1                                   

    # function for inserting an element 
    ## we need to shift position of each and every element from the postion to insert.
    def insert(self, keyToInsert, position):
        if(self.sizeOfArray > position):
            for i in range(self.sizeOfArray - 2, position - 1, -1):
                self.arrayItems[i + 1] = self.arrayItems[i]
            self.arrayItems[position] = keyToInsert
        else:
            print('Array size is:', self.sizeOfArray)

    # function to delete an element
    def delete(self, keyToDelete, position):
        if(self.sizeOfArray > position):
            for i in range(position, self.sizeOfArray - 1):
                self.arrayItems[i] = self.arrayItems[i + 1]
            self.arrayItems[i + 1] = self.arrayType(0)
        else:
            print('Array size is:', self.sizeOfArray)

if __name__ == '__main__':
    a = Array(10, int)
    print(a)
    a.insert(2, 2)              ###-------> this operation requires shifting of every index whch again takes o(n)
    print(a)
    b=a.search(2)
    print(b)                    ### -------> returns the value 2 as it is present
    c=a.search(3)
    print(c)                    ###-------> returns -1 as value '3' is not present in the array
    a.insert(4,7)
    print(a)
    print(len(a))
```
